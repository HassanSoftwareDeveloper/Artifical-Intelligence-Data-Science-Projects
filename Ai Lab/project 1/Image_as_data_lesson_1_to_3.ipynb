Task 1.1.7: Use both the mathematical operator and the class method to multiply left_tensor to right_tensor. Assign the results to ew_tensor_operator and ew_tensor_method, respectively.

One of the most important mathematical operations in deep learning is multiplication, so let's spend some time on it here.

Keep in mind that, when it comes to tensors, there's more than one type of multiplication. For starters, there's element-wise multiplication, where the corresponding values of two tensors are multiplied together. In PyTorch, we can do this using the * operator or the .mul() method.

```python
ew_tensor_operator = left_tensor * right_tensor
ew_tensor_method = left_tensor.mul(right_tensor)

print("ew_tensor_operator class:", type(ew_tensor_operator))
print("ew_tensor_operator shape:", ew_tensor_operator.shape)
print("ew_tensor_operator data type:", ew_tensor_operator.dtype)
print("ew_tensor_operator device:", ew_tensor_operator.device)
print(ew_tensor_operator)
print()
print("ew_tensor_method class:", type(ew_tensor_method))
print("ew_tensor_method shape:", ew_tensor_method.shape)
print("ew_tensor_method data type:", ew_tensor_method.dtype)
print("ew_tensor_method device:", ew_tensor_method.device)
print(ew_tensor_method)
```

Next, there's matrix multiplication, which combines the rows and columns of two tensors to generate a new one. We can use the @ operator or the .matmul() method.

To see how this works, let's create two new tensors with different shapes.

```python
left_tensor * right_tensor == right_tensor * left_tensor
```

```python
new_left_tensor = torch.Tensor([[2, 5], [7, 3]])
new_right_tensor = torch.Tensor([[8], [9]])

print("new_left_tensor class:", type(new_left_tensor))
print("new_left_tensor shape:", new_left_tensor.shape)
print("new_left_tensor data type:", new_left_tensor.dtype)
print("new_left_tensor device:", new_left_tensor.device)
print(new_left_tensor)
print()
print("new_right_tensor class:", type(new_right_tensor))
print("new_right_tensor shape:", new_right_tensor.shape)
print("new_right_tensor data type:", new_right_tensor.dtype)
print("new_right_tensor device:", new_right_tensor.device)
print(new_right_tensor)
```

Task 1.1.8: Use both the mathematical operator and the class method to perform matrix multiplication on new_left_tensor and new_right_tensor. Assign the results to mm_tensor_operator and mm_tensor_method, respectively.

```python
mm_tensor_operator = new_left_tensor @ new_right_tensor
mm_tensor_method = new_left_tensor.matmul(new_right_tensor)

print("mm_tensor_operator class:", type(mm_tensor_operator))
print("mm_tensor_operator shape:", mm_tensor_operator.shape)
print("mm_tensor_operator data type:", mm_tensor_operator.dtype)
print("mm_tensor_operator device:", mm_tensor_operator.device)
print(mm_tensor_operator)
print()
print("mm_tensor_method class:", type(mm_tensor_method))
print("mm_tensor_method shape:", mm_tensor_method.shape)
print("mm_tensor_method data type:", mm_tensor_method.dtype)
print("mm_tensor_method device:", mm_tensor_method.device)
print(mm_tensor_method)
```

One very important thing to remember: matrix multiplication is not commutative. The number of columns in the tensor on the left must equal the number of rows in the tensor on the right. If these two dimensions don't match, your code will throw a RunTimeError.

                           *Preparing Our Data*
We need to make sure all of our images are in the same mode. We want to make sure they're in RGB format. If not, we'll need to convert them. In fact, our sample hog image was in grayscale, denoted by the "L" for its mode value. We'll create a custom transformer that will convert an image to RGB mode if it's not.

```python
class ConvertToRGB:
    def __call__(self, img):
        if img.mode != "RGB":
            img = img.convert("RGB")
        return img
```

**Task 1.3.5: Add the missing last step where we convert images to PyTorch tensors. In a previous lesson we used transform.ToTensor to accomplish the conversion.**

```python
# Define transformation to apply to the images
transform = transforms.Compose(
    [
        ConvertToRGB(),  # Convert images to RGB format if not already
        transforms.Resize((224, 224)),  # Resize images to 224x224
        # Convert images to tensors
        transforms.ToTensor()
    ]
)

print(type(transform))
print(transform)
```

# Training our model
There are several pieces we need beyond the model object. First, we need to define a loss function. The loss function measures how well our model does for a given set of model parameters. Now we'll create a function called train_epoch that encapsulates the training process. The code is daunting at first but we're going to break it down. The function accepts

**model:** The PyTorch model we built with a specific architecture

**optimizer:** The optimizer that will be used to best adjust the model weights

**loss_fn:** The loss function that the optimizer is trying to minimize

**data_loader:** The DataLoader object for the training dataset that makes it easy to iterate over batches

**device:**The device where we're going to place the tensors

The function returns the average loss function on the training data over the entire epoch. Later, we'll use this to see if the model is making progress in the training process.

```python
from tqdm import tqdm  # for progress bar

def train_epoch(model, optimizer, loss_fn, data_loader, device="cpu"):
    # Start with 0 total loss — we'll add to this as we go
    total_loss = 0.0

    # Tell the model: "You’re in training mode now"
    model.train()

    # Go through the whole training set, one small batch at a time
    for inputs, targets in tqdm(data_loader, desc="Training", leave=False):
        # 1. Clear any leftover gradients (from the last batch)
        optimizer.zero_grad()

        # 2. Move input images and labels to the correct device (CPU or GPU)
        inputs = inputs.to(device)
        targets = targets.to(device)

        # 3. Make predictions (forward pass)
        output = model(inputs)

        # 4. Calculate how far off the predictions are (the loss)
        loss = loss_fn(output, targets)

        # 5. Figure out how to improve (backward pass)
        loss.backward()

        # 6. Update the model's weights to do better next time
        optimizer.step()

        # 7. Add this batch's loss (multiplied by batch size) to total
        total_loss += loss.item() * inputs.size(0)

    # 8. Return the average loss over the whole training set
    average_loss = total_loss / len(data_loader.dataset)
    return average_loss
```

The loss function we're using in this problem, the cross-entropy, is difficult for humans to interpret. We know that lower is better, but is that number good or bad? To help us humans judge, we'll calculate the accuracy of the model, the fraction of predictions it gets right. To do that, we need the model to make predictions.

The following function will make a prediction for each row of data in data_loader, using model. As with the train_epoch function above, it's a bit daunting. The comments explain what each section does.

```python
def predict (model, data_loader,device="cpu"):
  #start with an empty tensor to hold all the predictions(probailities)
  all_probs=torch.tensor([]).to(device)
  #tell the model you are in prediction mode now
  # This turn off things like dropout that are only used during trainning
  model.eval()
  #we are not trainning so we dont need gradients
  #This makes the code faster and saves memory
  with torch.no_grad():
    #go through each small batch of data
    for inputs, targets in tqdm (data_loader,desc="Predicting", leave=False):
      #Move input data to the same device as the model (cpu or gpu)
      inputs=inputs.to(device)
      #do a forward pass get the raw outputs (logits) from the model
      output=model(inputs)
      # convert the raw outputs a proabilities using softmax
      probs=F.softmax(output,dim=1)

      #Add these proabilities to our big list of all predictions
      all_probs = torch.cat((all_probs, probs),dim=0)
  #return the final list of all predicted proabilites
  return all_probs
```

```python
from tqdm import tqdm  # For progress bar

def score(model, data_loader, loss_fn, device="cpu"):
    # Start with 0 loss and 0 correct predictions
    total_loss = 0
    total_correct = 0

    # Put the model in "evaluation mode"
    # This turns off layers like dropout that behave differently during training
    model.eval()

    # We don't need gradients while testing — this makes things faster
    with torch.no_grad():
        # Loop through the data in batches
        for inputs, targets in tqdm(data_loader, desc="Scoring", leave=False):
            # Move inputs and labels to the same device as the model (CPU or GPU)
            inputs = inputs.to(device)
            targets = targets.to(device)

            # Make predictions (forward pass)
            output = model(inputs)

            # Compute the loss for this batch
            loss = loss_fn(output, targets)

            # Add this batch's loss to the total (multiply by batch size)
            total_loss += loss.item() * inputs.size(0)

            # Get the predicted class by choosing the one with the highest score
            predictions = torch.argmax(output, dim=1)

            # Check how many predictions were correct
            correct = torch.eq(predictions, targets)

            # Add the number of correct predictions to the total
            total_correct += torch.sum(correct).item()

    # Calculate the average loss and accuracy over the full dataset
    average_loss = total_loss / len(data_loader.dataset)
    accuracy = total_correct / len(data_loader.dataset)

    return average_loss, accuracy
```

We now have the pieces we need for our training loop. This loop will run over a number of epochs, calling the train_epoch function and the score function, and printing out some diagnostic information so we can see how the training is going.

```python
def train(model, optimizer, loss_fn, train_loader, val_loader, epochs=20, device="cpu"):
    # Loop through the number of epochs (training rounds)
    for epoch in range(1, epochs + 1):

        # 🔁 1. Train the model for one full pass over the training data
        training_loss = train_epoch(model, optimizer, loss_fn, train_loader, device)

        # ✅ 2. Evaluate the model on the validation data
        validation_loss, validation_accuracy = score(model, val_loader, loss_fn, device)

        # 📢 3. Print the results for this epoch
        print(
            f"Epoch: {epoch}, Training Loss: {training_loss:.2f}, "
            f"Validation Loss: {validation_loss:.2f}, Validation Accuracy: {validation_accuracy:.2f}"
        )
```
